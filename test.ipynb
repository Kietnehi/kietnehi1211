```python
import os
from crewai import Agent, Task, Crew, Process
from langchain_community.tools import DuckDuckGoSearchRun

# 1. C·∫•u h√¨nh c√¥ng c·ª• t√¨m ki·∫øm
search_tool = DuckDuckGoSearchRun()

# 2. ƒê·ªãnh nghƒ©a c√°c Agents
# Agent 1: T√¨m ki·∫øm th√¥ng tin
researcher = Agent(
    role='Chuy√™n gia ph√¢n t√≠ch th·ªã tr∆∞·ªùng',
    goal='T√¨m ki·∫øm nh·ªØng th√¥ng tin m·ªõi nh·∫•t v·ªÅ {topic} b·∫±ng DuckDuckGo',
    backstory='B·∫°n l√† m·ªôt chuy√™n gia sƒÉn tin, c√≥ kh·∫£ nƒÉng t√¨m ra nh·ªØng xu h∆∞·ªõng m·ªõi nh·∫•t v√† d·ªØ li·ªáu ch√≠nh x√°c nh·∫•t.',
    tools=[search_tool],
    verbose=True, # ƒê·ªÉ b·∫°n th·∫•y Agent suy nghƒ© v√† l√†m vi·ªác
    allow_delegation=False
)

# Agent 2: Vi·∫øt n·ªôi dung
writer = Agent(
    role='Bi√™n t·∫≠p vi√™n n·ªôi dung',
    goal='Vi·∫øt m·ªôt b√†i t√≥m t·∫Øt ng·∫Øn g·ªçn v√† s√∫c t√≠ch v·ªÅ {topic} d·ª±a tr√™n d·ªØ li·ªáu t√¨m ƒë∆∞·ª£c',
    backstory='B·∫°n l√† m·ªôt c√¢y vi·∫øt xu·∫•t s·∫Øc, c√≥ kh·∫£ nƒÉng bi·∫øn nh·ªØng d·ªØ li·ªáu kh√¥ khan th√†nh b√†i vi·∫øt h·∫•p d·∫´n.',
    verbose=True,
    allow_delegation=False
)

# 3. ƒê·ªãnh nghƒ©a c√°c nhi·ªám v·ª• (Tasks)
task1 = Task(
    description='S·ª≠ d·ª•ng DuckDuckGo ƒë·ªÉ t√¨m ki·∫øm 3 tin t·ª©c quan tr·ªçng nh·∫•t v·ªÅ {topic} trong tu·∫ßn n√†y.',
    expected_output='Danh s√°ch 3 tin t·ª©c d∆∞·ªõi d·∫°ng g·∫°ch ƒë·∫ßu d√≤ng k√®m ngu·ªìn.',
    agent=researcher
)

task2 = Task(
    description='D·ª±a tr√™n th√¥ng tin t·ª´ nh√† nghi√™n c·ª©u, h√£y vi·∫øt m·ªôt b·∫£n tin ng·∫Øn kho·∫£ng 150 ch·ªØ v·ªÅ {topic}.',
    expected_output='M·ªôt b√†i t√≥m t·∫Øt ho√†n ch·ªânh b·∫±ng ti·∫øng Vi·ªát.',
    agent=writer
)

# 4. Th√†nh l·∫≠p "ƒê·ªôi ng≈©" (Crew)
my_crew = Crew(
    agents=[researcher, writer],
    tasks=[task1, task2],
    process=Process.sequential # L√†m tu·∫ßn t·ª±: Researcher l√†m xong r·ªìi m·ªõi ƒë·∫øn Writer
)

# 5. B·∫Øt ƒë·∫ßu ch·∫°y
print("### ƒêang b·∫Øt ƒë·∫ßu ch·∫°y CrewAI...")
result = my_crew.kickoff(inputs={'topic': 'Tr√≠ tu·ªá nh√¢n t·∫°o Agentic AI nƒÉm 2024'})

print("\n\n########################")
print("## K·∫æT QU·∫¢ CU·ªêI C√ôNG ##")
print("########################\n")
print(result)
```
